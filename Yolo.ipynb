```python
#importa√ß√£o da biblioteca e carregamento do modelo
from ultralytics import YOLO

model = YOLO("yolov8n-cls.pt")
```

```python
#caminho do dataset
caminho = "/home/lucas/Documentos/Lamia/ProjetoFinal/data/dataset_eye_disease"

results = model.train(
    data=caminho,
    epochs=100, #qnt de epocas
    imgsz=224, #tamanho da imagem
    augment=True,  #augment 

    #transforma√ß√µes leves do augmente
    degrees=20.0,         
    translate=0.2,        
    scale=0.6,            
    shear=5.0,            
    hsv_s=0.8,            
    hsv_v=0.5,            
    flipud=0.1,           
    mixup=0.15,          
    erasing=0.5           
)
```
